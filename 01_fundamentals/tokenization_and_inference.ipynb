## 1. Tokenization

LLMs don't read words — they read **tokens** (chunks of characters). Understanding tokens helps you:
- Know how much text fits in the context window
- Debug unexpected model behavior
- Control costs when using API-based models


```python
from transformers import AutoTokenizer

# Load a tokenizer — we use a small model for speed
tokenizer = AutoTokenizer.from_pretrained('distilgpt2')

text = 'Large language models learn patterns from vast amounts of text data.'
tokens = tokenizer.encode(text)

print(f'Text: {text}')
print(f'Number of tokens: {len(tokens)}')
print(f'Token IDs: {tokens}')

# Decode back to see individual tokens
for token_id in tokens:
    print(f'  {token_id:6d} -> {repr(tokenizer.decode([token_id]))}')
```

## 2. Context Window

Every model has a **context window** — the maximum number of tokens it can process at once.
Text outside this window is invisible to the model. This is why context management matters!


```python
print(f'distilgpt2 max context length: {tokenizer.model_max_length} tokens')

# Rule of thumb: 1 token ≈ 0.75 words
max_words_approx = tokenizer.model_max_length * 0.75
print(f'Roughly {max_words_approx:.0f} words fit in the context window')
```

## 3. Loading a Model and Generating Text


```python
from transformers import AutoModelForCausalLM, pipeline
import torch

model_name = 'distilgpt2'

# pipeline() is the simplest way to run inference
generator = pipeline(
    'text-generation',
    model=model_name,
    device='cpu'  # change to 'cuda' if you have a GPU
)

prompt = 'The key to maintaining an accurate AI system is'
result = generator(
    prompt,
    max_new_tokens=50,
    num_return_sequences=1,
    do_sample=True,
    temperature=0.7
)

print(result[0]['generated_text'])
```

## 4. Key Parameters Explained

| Parameter | What it controls |
|-----------|------------------|
| `max_new_tokens` | How many tokens to generate |
| `temperature` | Randomness (0 = deterministic, 1 = creative, >1 = chaotic) |
| `top_p` | Nucleus sampling — limits token pool to top P probability mass |
| `do_sample` | If False, always picks the most likely token (greedy) |

**Exercise:** Try changing `temperature` between 0.1 and 1.5 and observe how the output changes.


```python
for temp in [0.1, 0.5, 1.0, 1.5]:
    result = generator(
        prompt,
        max_new_tokens=30,
        do_sample=True,
        temperature=temp
    )
    print(f'\nTemperature {temp}:')
    print(result[0]['generated_text'])
```
